## Perlin Noise

```python
import noise
import numpy as np

# Lacunarity: number that determines how much detail is added or removed at each octave (adjusts frequency). 
# Persistence: number that determines how much each octave contributes to the overall shape (adjusts amplitude).

shape = (50,50)
scale = 11
elevation_map = np.zeros(shape)
for i in range(shape[0]):
    for j in range(shape[1]):
        elevation_map[i,j] = noise.pnoise2(i/scale, j/scale, octaves=1, persistence=0.15, lacunarity=1, repeatx=1024, repeaty=1024, base=0)
print(elevation_map)    
```

```python
from matplotlib import pyplot as plt

plt.imshow(elevation_map,cmap='gray')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

data = elevation_map
fig, (ax, bx) = plt.subplots(nrows=1, ncols=2, num=0, figsize=(16, 8),
                             subplot_kw={'projection': '3d'})
for i in range(data.shape[1]):
    ax.plot3D(np.repeat(i, data.shape[0]), np.arange(data.shape[0]), data[:, i])
gridY, gridX = np.mgrid[1:data.shape[0]:data.shape[0] * 1j,
                           1:data.shape[1]:data.shape[1] * 1j]
pSurf = bx.plot_surface(gridX, gridY, data, cmap='viridis')
fig.colorbar(pSurf)
plt.show()
```
